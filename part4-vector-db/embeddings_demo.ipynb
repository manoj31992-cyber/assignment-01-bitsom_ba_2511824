```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

model = SentenceTransformer('all-MiniLM-L6-v2')
print('Model loaded.')
```

```python
# 10 sentences across 3 topics (at least 3 per topic)
sentences = [
    # Cricket (indices 0–3)
    'The batsman hit a powerful six over mid-wicket.',
    'India won the Test series against Australia by three wickets.',
    'The spinner bowled a perfect googly that deceived the batsman.',
    'The fielder took a stunning catch at the boundary rope.',
    # Cooking (indices 4–6)
    'Sauté the onions in olive oil until they turn golden brown.',
    'Always preheat the oven to 180 degrees before baking a cake.',
    'Add a pinch of salt and freshly ground pepper to season the soup.',
    # Cybersecurity (indices 7–9)
    'A SQL injection attack can expose sensitive database records.',
    'Always use multi-factor authentication to protect online accounts.',
    'Ransomware encrypts your files and demands payment for decryption.',
]

labels = [
    'Cricket-1','Cricket-2','Cricket-3','Cricket-4',
    'Cooking-1','Cooking-2','Cooking-3',
    'CyberSec-1','CyberSec-2','CyberSec-3',
]

embeddings = model.encode(sentences)
print(f'Embeddings shape: {embeddings.shape}')
```

```python
# Compute 10x10 cosine similarity matrix
sim_matrix = cosine_similarity(embeddings)

# Plot heatmap
plt.figure(figsize=(11, 9))
sns.heatmap(
    sim_matrix,
    xticklabels=labels,
    yticklabels=labels,
    annot=True,
    fmt='.2f',
    cmap='YlOrRd',
    vmin=0, vmax=1,
    linewidths=0.5
)
plt.title('10×10 Cosine Similarity Matrix\n(Cricket | Cooking | Cybersecurity)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('similarity_heatmap.png', dpi=120)
plt.show()
print('Heatmap saved.')
```

```python
# Query: find top-2 most similar sentences
query = 'The bowler took three wickets in one over'
query_embedding = model.encode([query])

scores = cosine_similarity(query_embedding, embeddings)[0]
top2_idx = np.argsort(scores)[::-1][:2]

print(f'Query: "{query}"\n')
print('Top 2 most similar sentences:')
for rank, idx in enumerate(top2_idx, 1):
    print(f'  {rank}. [{labels[idx]}] "{sentences[idx]}"')
    print(f'     Similarity score: {scores[idx]:.4f}')
```
